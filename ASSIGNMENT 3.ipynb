# which Workclass getting the Highest Salary?

```python
import pandas as pd
data = {'workclass': ['Private', 'Public', 'Private', 'Self-employed', 'Public'],
        'fnlwgt': [50000, 60000, 55000, 70000, 65000]}


df = pd.DataFrame(data)
df['workclass'] = df['workclass'].astype('category')
for workclass, group in df.groupby('workclass'):
    print(f"Workclass: {workclass}")
    print(group['fnlwgt'].mean())
```

# how has better Chance to get Salary>50k Male or Female

```python
cross_tab = pd.crosstab(data['gender'], data['income'], margins=True, margins_name='Total')
cross_tab['Percentage >50k'] = (cross_tab['>50K'] / cross_tab['Total']) * 100
print(cross_tab[['Percentage >50k']])
```
