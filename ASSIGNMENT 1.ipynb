#         # how many people have French 'fr' as their Language

```python
#Import the data
import pandas as pd
import matplotlib as plt
import seaborn as sns
data=pd.read_csv("Ecommerce_Purchases")
#Count french speakers
french_speakers = data['Language'].value_counts().get('fr', 0)
print(f'Number of people with French as their language: {french_speakers}')
# Plot the bar graph
sns.countplot(x='Language', data=data)
plt.pyplot.title('Number of People Speaking Each Language')
plt.pyplot.xlabel('Language')
plt.pyplot.ylabel('Number of People')
plt.pyplot.show()
```

# job title contains engineer

```python
#engineer_count = df[df['Job'].str.contains('engineer', case=False, na=False)].shape[0]
#print(f'The number of people with "engineer" in their job title: {engineer_count}')
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have already read the CSV file into a DataFrame
data_ecommerce = pd.read_csv("Ecommerce Purchases")

# Count the occurrences of each job title
job_title_counts = data_ecommerce['Job'].value_counts()

# Plot a bar graph
plt.figure(figsize=(14, 6))
sns.countplot(x='Job', data=data_ecommerce[data_ecommerce['Job'].str.contains('Engineer', case=False)], palette='viridis')
plt.title('Distribution of Job containing "Engineer" in Ecommerce Purchases')
plt.xlabel('Job')
plt.ylabel('Number of People')
plt.xticks(rotation=45, ha='right')
plt.show()

# Display the count of people with 'Engineer' in their job title
engineer_count = job_title_counts.get('Engineer', 0)
print(f'The number of people with "Engineer" in their job: {engineer_count}')
```

# how many people have Mastercard as their Credit Card Provider and Made A purchase above 50?

# Find Email of the person with the following Credit Card number:4105595335494659

```python
import pandas as pd
df = pd.read_csv('Ecommerce Purchases')
credit_card_number_to_search = '4105595335494659'
if any(df['Credit Card'] == credit_card_number_to_search):
    email_of_person = df.loc[df['Credit Card'] == credit_card_number_to_search, 'Email'].iloc[0]
    print(f'The email of the person with Credit Card Number {credit_card_number_to_search}: {email_of_person}')
else:
    print(f'No record found for Credit Card Number {credit_card_number_to_search}')
```

# how Many people Purchase During The Am and How Many People Purchase During Pm?

```python
am_purchase_count = df[df['AM or PM'].str.contains('AM')].shape[0]
pm_purchase_count = df[df['AM or PM'].str.contains('PM')].shape[0]
print(f'Number of people who made purchases during AM: {am_purchase_count}')
print(f'Number of people who made purchases during PM: {pm_purchase_count}')
```
